```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,5,10)
ax.plot(x,np.sin(x));
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,53,100)
print(x)
ax.plot(x,np.sin(x));
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,53,100)
print(x)
ax.plot(x,np.cos(x));
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.cos(x));
ax.plot(x,np.sin(x));
ax.plot(x,np.tan(x));
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.cos(x), label='cos',color='red')
ax.plot(x,np.sin(x), label='sin',color='green')
ax.set_title('mupltiple list')
ax.set_xlabel('preiod')
ax.set_ylabel('amplitute')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.cos(x), label='cos',color='red',linestyle=('dashed'))
ax.plot(x,np.sin(x), label='sin',color='green',linestyle=('dotted'))
ax.plot(x,np.sin(x+1), label='sin x+1',color='green',linestyle=('dotted'))
ax.set_title('mupltiple list')
ax.set_xlabel('preiod')
ax.set_ylabel('amplitute')
ax.legend()
plt.show()
```
